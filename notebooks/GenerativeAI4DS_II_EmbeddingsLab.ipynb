```python
from langchain_openai import OpenAIEmbeddings
```

### 1. SetUp


```python
## We rely on OpenAI to create embeddings
embeddings_model = OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)
```

### 2. Document embedding

```python
document="My new electric car is awesome, it is silent, easy to drive and saves me lots of money"

```

```python

embedded_document = embeddings_model.embed_query(document)
```

```python

type(embedded_document)
```

```python
print(embedded_document)
```

### 3. List of Documents embedding

```python
conversations=[
        "Hi there!",
        "Oh, hello!",
        "What's your name?",
        "My friends call me World",
        "Hello World!"
    ]
```

```python
conversations
```

```python
embedded_conversations = embeddings_model.embed_documents(conversations)
```

```python
type(embedded_conversations)
```

```python
print(embedded_conversations)
```
